```python
# Loading data
df_info = pd.read_csv('Data//superhero_info - superhero_info.csv')
df_powers = pd.read_csv('Data//superhero_powers - superhero_powers.csv')
```

```python
## save the 2 new columns into the dataframe
df_info[['Hero','Publisher']] = df_info['Hero|Publisher'].str.split('|',expand=True)
df_info.head(3)
```

```python
## drop the original column
df_info = df_info.drop(columns=['Hero|Publisher'])
df_info.head(3)
```

```python
# measurements column
df_info['Measurements'].head(5)
```

```python
# split Measuremnts column into 2 column Height and Weight
df_info[['Height(cm)', 'Weight(kg)']] = df_info['Measurements'].str.split(',', expand=True)
df_info.head()
```

```python
# dropping the Measurements column.
df_info = df_info.drop(columns=['Measurements'])
df_info.head(3)
```

```python
# Make a list of all characters to replace
to_replace = ['{','}', "'Height': ", "'", ' cm']
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    df_info['Height(cm)'] = df_info['Height(cm)'].str.replace(char,'',regex=False)

# Remove 'Height: ' from the beginning of the values
df_info['Height(cm)'] = df_info['Height(cm)'].str.replace('Height: ', '', regex=False)

df_info['Height(cm)'].head()

```

```python
# Make a list of all characters to replace
to_replace = ['{', '}', "'Weight': ", "'", ' kg']
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    df_info['Weight(kg)'] = df_info['Weight(kg)'].str.replace(char,'',regex=False)

# Remove 'Weight: ' from the beginning of the values
df_info['Weight(kg)'] = df_info['Weight(kg)'].str.replace('Weight: ', '', regex=False)

df_info['Weight(kg)'].head()

```

```python
df_info['Height(cm)'] = df_info['Height(cm)'].str.replace("'",'')
df_info.head(3)
```

```python
df_info['Weight(kg)'] = df_info['Weight(kg)'].str.replace("'",'')
df_info['Weight(kg)'].head()
```

```python
# convert weight and height to float(decimal)
df_info['Height(cm)'] = df_info['Height(cm)'].astype(float)
df_info['Weight(kg)'] = df_info['Weight(kg)'].astype(float)
print(df_info.dtypes)
```

```python
# Reorder the columns as per your request
df_cleaned_info = df_info[["Hero", "Publisher", "Gender", "Eye color", "Race", "Hair color", "Height(cm)", "Skin color", "Alignment", "Weight(kg)"]]
df_cleaned_info
```

```python
# Ensure the 'Powers' column is a string type
df_powers['Powers'] = df_powers['Powers'].astype(str)

# Split the 'Powers' column into separate columns
powers_split = df_powers['Powers'].str.split(',', expand=True)

# Create one-hot-encoded columns for each unique power
unique_powers = powers_split.stack().str.get_dummies().max(level=0)

# Concatenate the one-hot-encoded columns with the original DataFrame
df_powers = pd.concat([df_powers, unique_powers], axis=1)

# Drop the original 'Powers' column
df_powers.drop('Powers', axis=1, inplace=True)
```

```python
df_powers.head()
```

```python
#merging the two files
df_clean_data = df_cleaned_info.merge(df_powers, how='cross')
df_clean_data
```

```python
# new dataframe
```

```python
# Calculate the average weight of characters with Super Speed
average_weight = df_clean_data[df_clean_data['Super Speed'] == 1]['Weight(kg)'].mean()

# Calculate the average weight of characters without Super Speed
average_weight = df_clean_data[df_clean_data['Super Speed'] == 0]['Weight(kg)'].mean()

print("Average weight of characters with Super Speed:", average_weight)
print("Average weight of characters without Super Speed:", average_weight)
```

```python
# Group the DataFrame by 'Publisher' and calculate the average height for each group
average_height = df_clean_data.groupby('Publisher')['Height(cm)'].mean()

# Display the result
print(average_height)
```
